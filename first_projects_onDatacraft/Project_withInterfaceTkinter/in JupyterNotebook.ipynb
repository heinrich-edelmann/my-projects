```python
import tkinter as tk
import tkinter.messagebox


window = tk.Tk()
window.geometry("500x200")
window.title("Hauspreisrechner")

#Erstellung meiner Arbeitswindow:
labelListboxGrundstück = tk.Label(window, text="Grundstück in m2")
Grundstück_m2 = tk.Entry(window)
labelListboxWohnfläche = tk.Label(window, text="Wohnfläche in m2")
Wohnfläche_m2 = tk.Entry(window)
labelListboxBaujahr = tk.Label(window, text="Baujahr")
Baujahr = tk.Entry(window)
ausgabe = tk.Label(window)


#Tabellen: 
BundK # Bundesland und Kostenfaktor
SLundK # StadtvsLand und Kostenfaktor
AundK # Ausstattung und Kostenfaktor
HundK # Hausart und Kostenfaktor
WKproPE # Weitere Kostenfaktoren pro Preis und Einheit


# meine Berechnungen:

preis_fallsMakler = float(WKproPE.loc["Makler"])*0.01 # → mal 0.01 um es in % zu konvertieren
preis_fallsDenkmalschutz = float(WKproPE.loc["Denkmalschutz"])*0.01
preis_QM_Grunstück_nachTabelle = float(WKproPE.loc["QM-Grundstück"])   
preis_QM_Wohnfläche_nachTabelle = float(WKproPE.loc["QM-Wohnfläche"])   
    
#-----------------------------------------------------------------------------------------------------------------------------------------------
def berechnen():
    bdl = listbox_Bundesland.get(listbox_Bundesland.curselection()) # → .get() gibt immer ein String zurück
    slk = listbox_StadtvsLand.get(listbox_StadtvsLand.curselection())# mylistbox.curselection private function von get bei Listbox
    ak = listbox_Ausstattung.get(listbox_Ausstattung.curselection())
    ha = listbox_Hausart.get(listbox_Hausart.curselection())
    Gm2 = float(Grundstück_m2.get())
    berechnung_Grundstück = Gm2 * preis_QM_Grunstück_nachTabelle
    WohnFm2 = float(Wohnfläche_m2.get()) 
    berechnung_Wohnfläche = WohnFm2 * preis_QM_Wohnfläche_nachTabelle
    Baujahrkosten = (float(Baujahr.get()) - 2022) * (float(WKproPE.loc["Baujahr"])/100)    
    flächegesamt = berechnung_Grundstück + berechnung_Wohnfläche - Baujahrkosten 
    ergebnis = round(flächegesamt * float(BundK.loc[bdl]) * float(SLundK.loc[slk]) * float(AundK.loc[ak]) * float(HundK.loc[ha]))
    
    def mitMakler_oderOhne(x):
        if var_Makler.get() == 1:
            mitMakler = x*preis_fallsMakler
            return mitMakler
        else:
            mitMakler = 0
            return mitMakler
    def mitDSchutz_oderOhne(y):
        if var_Makler.get() == 1:
            mitDenkmalschutz = y*preis_fallsDenkmalschutz
            return mitDenkmalschutz
        else:
            mitDenkmalschutz = 0
            return mitDenkmalschutz
        
    ende_ergebnis = ergebnis + mitMakler_oderOhne(ergebnis) + mitDSchutz_oderOhne(ergebnis)
    ausgabe.configure(text=f"ihr zukunftiges Haus wird {ende_ergebnis}€ kosten")
#------------------------------------------------------------------------------------------------------------------------------------------------

#meine Variabeln für meine Buttons:
var_Makler = tk.IntVar()
var_Denkmalschutz = tk.IntVar()
var_Bundesland = tk.Variable(value = list(BundK.index))
var_StadtvsLand = tk.Variable(value = list(SLundK.index))
var_Ausstattung = tk.Variable(value = list(AundK.index))
var_Hausart = tk.Variable(value = list(HundK.index))
var_weitereKosten = tk.Variable(value = list(WKproPE.index))


# o que eu escolho passa entao para a variavel listbox_Bundesland zB, valor esse determinado pelos itens 
#q eu add à Listbox através do meu listvariable, que estao em var_Bundesland

def resetFunktion():
    var_Makler.set(0)
    var_Denkmalschutz.set(0)




#Erstellung meiner Buttons:
labelListboxBundesland =tk.Label(window, text="Wählen Sie bitte ein Bundesland aus")
listbox_Bundesland = tk.Listbox(window, listvariable = var_Bundesland, height=2, selectmode=tk.EXTENDED, exportselection = 0)

labelListboxStadtvsLand =tk.Label(window, text="Wählen Sie die Wohnortart")
listbox_StadtvsLand = tk.Listbox(window, listvariable = var_StadtvsLand, height=1, selectmode=tk.EXTENDED, exportselection = 0)

labelListboxAusstattung =tk.Label(window, text="Wählen Sie bitte die Ausstattungsform aus")
listbox_Ausstattung = tk.Listbox(window, listvariable = var_Ausstattung, height=2, selectmode=tk.EXTENDED, exportselection = 0)

labelListboxHausart =tk.Label(window, text="Wählen Sie bitte die Hausart aus")
listbox_Hausart = tk.Listbox(window, listvariable = var_Hausart, height=2, selectmode=tk.EXTENDED, exportselection = 0)

Radiobutton_Makler = tk.Radiobutton(window, text="Makler", variable = var_Makler, value = 1)                            
Radiobutton_Denkmalschutz = tk.Radiobutton(window, text="Denkmalschutz", variable = var_Denkmalschutz, value = 1)

ErgebnisButton = tk.Button(window, text="Hauspreis", command = berechnen)
ResetButton = tk.Button(window, text="Reset Makler und Denkmalschutz", command=resetFunktion)

# alles ins Window
labelListboxGrundstück.pack()
Grundstück_m2.pack()
labelListboxWohnfläche.pack()
Wohnfläche_m2.pack()
labelListboxBaujahr.pack()
Baujahr.pack()
ausgabe.pack()

labelListboxBundesland.pack()
listbox_Bundesland.pack(expand=True, fill=tk.BOTH)

labelListboxStadtvsLand.pack()
listbox_StadtvsLand.pack(expand=True, fill=tk.BOTH)

labelListboxAusstattung.pack()
listbox_Ausstattung.pack(expand=True, fill=tk.BOTH)

labelListboxHausart.pack()
listbox_Hausart.pack(expand=True, fill=tk.BOTH)

Radiobutton_Makler.pack()
Radiobutton_Denkmalschutz.pack()

ErgebnisButton.pack()
ResetButton.pack()

#Los!
tk.mainloop()
```
