### Wie viele Filme pro Jahrzehnte gibts in der Tabelle

```python
lieblingsJahrgang = Top250list["Date"].value_counts() 
lieblingsJahrgang
```

```python
lji = list(lieblingsJahrgang.index)
print(lji)
```

```python
zweitausendzehner = 0
zweitausender = 0
neunziger = 0
achtziger = 0
siebziger = 0
sechziger = 0
fünfziger = 0
vierziger = 0
dreiziger = 0
zwanziger = 0
for i in list(lji):
    if i <= 2022 and i >= 2010:
        zweitausendzehner += 1
    elif i < 2010 and i >= 2000:
        zweitausender += 1
    elif i < 2000 and i >= 1990:
        neunziger += 1
    elif i < 1990 and i >= 1980:
        achtziger += 1
    elif i < 1980 and i >= 1970:
        siebziger += 1
    elif i < 1970 and i >= 1960:
        sechziger += 1
    elif i < 1960 and i >= 1950:
        fünfziger += 1
    elif i < 1950 and i >= 1940:
        vierziger += 1
    elif i < 1940 and i >= 1930:
        dreiziger += 1
    elif i < 1930 and i >= 1920:
        zwanziger += 1
dictJahre = {"zweitausendzehner":zweitausendzehner, "zweitausender":zweitausender,"neunziger":neunziger,"achtziger":achtziger, "siebziger":siebziger, "sechziger":sechziger, "fünfziger":fünfziger, "vierziger":vierziger, "dreiziger":dreiziger, "zwanziger":zwanziger}
fürGrafik = pd.Series(dictJahre)     
fürGrafik           
```

```python
fürGrafik.plot.bar(rot = 285) 
plt.title("Filmszene")
plt.xlabel("Ercheinungsdatum")
plt.ylabel("Anzahl an Filme")

```
